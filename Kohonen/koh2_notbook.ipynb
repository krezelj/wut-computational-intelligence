```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.kohonen import *
```

```python
data_mnist = pd.read_csv('data/mio/mnist_train.csv')
data_mnist = data_mnist.iloc[:1000]
```

```python
X_train = data_mnist.iloc[:, 1:].values / 255
Y_train = data_mnist.iloc[:, 0].values
```

```python
koh = Kohonen(10, 10, 784, 7, mexican_hat)
koh.fit(X_train, 10)
```

```python
vertex_labels = koh.match(X_train)
```

```python
pos = np.zeros(shape=(koh.m * koh.n, 2))
for i in range(koh.m):
    for j in range(koh.n):
        pos[i * koh.m + j, 0] = i
        pos[i * koh.m + j, 1] = j
```

```python
plt.figure(figsize=(8, 6))
fig, ax = plt.subplots()

scatter = ax.scatter(x=pos[:,0], y=pos[:,1], c=(Y_train[vertex_labels]), s=500, cmap='tab10')
plt.xlim([-1, 12])
plt.ylim([-1, 10])

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Classes")
ax.add_artist(legend1)
```

```python
# sns.scatterplot(x=pos[:,0],y=pos[:,1], hue=Y_train[vertex_labels], s=250, palette='tab10', legend=False)
```

```python
koh = Kohonen(10, 10, 784, 7, mexican_hat, grid_type='hex')
koh.fit(X_train, 20)
```

```python
vertex_labels = koh.match(X_train)
```

```python
pos = np.zeros(shape=(koh.m * koh.n, 2))
for i in range(koh.m):
    for j in range(koh.n):
        pos[i * koh.m + j, 0] = i + (0 if j % 2 ==0 else 0.5)
        pos[i * koh.m + j, 1] = j
```

```python
plt.figure(figsize=(8, 6))
fig, ax = plt.subplots()

scatter = ax.scatter(x=pos[:,0], y=pos[:,1], c=(Y_train[vertex_labels]), s=800, marker='h', cmap='tab10')
plt.xlim([-1, 12])
plt.ylim([-1, 10])

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Classes")
ax.add_artist(legend1)
```

```python
X_train = pd.read_csv('data/mio/smartphone_activity/activity_X_train.csv', sep=';').values
Y_train = pd.read_csv('data/mio/smartphone_activity/activity_Y_train.csv', sep=';').values.reshape(-1)
```
