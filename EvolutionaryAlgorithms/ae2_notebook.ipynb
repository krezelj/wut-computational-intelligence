### Circle Cutting

```python
def circle_cutting(R, generations=20):
    CircleCutting.use_file(f'data/cutting/r{R}.csv')
    evolution = Evolution(CircleCutting, 100, 10, init_args={'radius': R, 'init_rectangles': 5}, allow_crossover=False)
    evolution.simulate(generations)
    s, e = evolution.get_best_specimen()
    while input(f"Current best fitness is {e}. Continue? [y/n] ") == 'y':
        generations = int(input("How many generations? "))
        evolution.simulate(generations)
        s, e = evolution.get_best_specimen()
    return evolution

def show_best_specimen(evolution):
    s, e = evolution.get_best_specimen()
    img = s.get_image()
    plt.imshow(img)
```

**R=800** (30.000)

```python
evolution_800 = circle_cutting(800, 80)
```

```python
_, e = evolution_800.get_best_specimen()
print(e)
```

```python
show_best_specimen(evolution_800)
```
