```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import mean_squared_error
from src.mlp import MLP, Layer
```

## Squares Simple

```python
df_training = pd.read_csv("data/mio1/regression/square-simple-training.csv", index_col=0)
df_test = pd.read_csv("data/mio1/regression/square-simple-test.csv", index_col=0)
```

```python
x_train = df_training['x'].values.reshape(1, 100)
y_train = df_training['y'].values.reshape(1, 100)
```

```python
sns.scatterplot(data=df_training, x="x", y="y")
```

```python
W1 = np.array([
    [-3.2], [-4.5], [3.2], [4.5], [0]
])
B1 = np.array([
    -3, -7.62, -3, -7.62, 0
]).reshape(5, -1)
W2 = np.array([
    [170, 242, 170, 242, 0]
])
B2 = np.array([
    -145
]).reshape(1, -1)

model = MLP(layers=[
    Layer(1, 5, W1, B1),
    Layer(5, 1, W2, B2, activation="linear")
])
```

```python
model.plot(log_weights=True)
```

```python
x_test = df_test['x'].values.reshape(1, 100)
y_test = df_test['y'].values.reshape(1, 100)

y_predicted = model.predict(x_test)
mean_squared_error(y_predicted, y_test)
```

```python
sns.scatterplot(x=x_test.reshape(100), y=y_predicted.reshape(100))
sns.scatterplot(x=x_test.reshape(100), y=y_test.reshape(100))
```
